** Going to process categorical columns and numerical separately. Going to treat categoricals different. **

```python
# left most bit corresponds to the '1' in '0b00000001'
# 0th corresponds to the right most digit and increases to the left
def convertHeroToEncoding(heroId, prefix):
    # this is the reverse of bin(128)[2:]
    # return in reverse order for convenience
    heroId = int(heroId)
    digits = bin(heroId)[2:]
    numberDigits = len(digits)
    digits = digits[::-1]
    data = {}
    for i in range(numberDigits):
        data[prefix + '_heroBit_' + str(i)] = digits[i]
    # don't need the 7th bit (left most) because it is always zero as hero IDs do not go over 127 (remember 8 digits total)
    for i in range(numberDigits, 7):
        data[prefix + '_heroBit_' + str(i)] = 0
    return data
```

```python
ml = []
for e in data['invokerItems']:
    ml = ml + [int(i) for i in e[1:-1].split(',')]
```

```python
highestItemID = max(ml)
```

```python
data['invokerItems'] = data['invokerItems'].apply(lambda x: x[1:-1].split(','))
```

```python
def encodeItem(itemId):
    return [int(i==int(itemId)) for i in range(highestItemID+1)]
```

```python
invokerItems = data['invokerItems'].apply(lambda x: [encodeItem(i) for i in x])
```

```python
invokerItems = invokerItems.values
```

```python
for i in range(len(invokerItems)):
    subarray = invokerItems[i]
    for j in range(len(subarray)):
        subarray[j] = np.array(subarray[j])
    invokerItems[i] = np.array(invokerItems[i])
```

```python
if not os.path.exists("invokerItems.npy"):
    np.save('invokerItems', invokerItems)
else:
    invokerItems = np.load('invokerItems.npy')
```

```python
data['invokerItems'] = invokerItems
```

**Now to get the rest of the features and process them**

```python
data.columns
```

```python
numericalFeatures = ['kills', 'deaths', 'assists', 'lastHits', 'denies', 'gold', 'xp', 'heroDamage', 'heroHealing']
__actualNumericalFeatures = [] 
for numFeature in numericalFeatures:
    for column in data.columns:
        if numFeature in column:
            __actualNumericalFeatures.append(column)
numericalFeatures = __actualNumericalFeatures
```

```python
def maxMinColumn(colv):
    maxh = np.nanmax(colv)
    minh = np.nanmin(colv)
    return (colv - minh) / (maxh - minh)
```

```python
for feature in numericalFeatures:
    data[feature] = maxMinColumn(data[feature])
```

```python
data.columns
```

```python
def injectHeroEncoding(row):
    data = {}
    data['matchId'] = int(row['matchId'])
    for ix in range(1,6):
        incDataOS = convertHeroToEncoding(prefix='OSHero'+str(ix), heroId=row['OSHero'+str(ix)+'hero'])
        if ix != 5:
            incDataIS = convertHeroToEncoding(prefix='ISHero'+str(ix), heroId=row['ISHero'+str(ix)+'hero'])
        for key in incDataOS:
            data[key] = incDataOS[key]
        for key in incDataIS:
            data[key] = incDataIS[key]
    return pd.Series(data)
```

```python
heroEncodingsDf = data.apply(injectHeroEncoding, axis=1)
data['matchId'] = data['matchId'].apply(lambda x: int(x))
```

```python
data = data.merge(heroEncodingsDf, left_on='matchId', right_on='matchId')
```

```python
listOfColumnstoDrop = []
for ix in range(1,6):
    listOfColumnstoDrop = ['OSHero'+str(ix)+'hero'] + listOfColumnstoDrop
    if ix != 5:
        listOfColumnstoDrop = ['ISHero'+str(ix)+'hero'] + listOfColumnstoDrop
data.drop(labels=listOfColumnstoDrop, inplace=True, axis=1)
```
